```python
import pandas as pd
import numpy as np

from google.colab import drive

from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

```python
train_path = "/content/drive/MyDrive/archive (3)/fraudTrain.csv"
test_path = "/content/drive/MyDrive/archive (3)/fraudTest.csv"

train_df = pd.read_csv(train_path)
test_df = pd.read_csv(test_path)

print("Train Shape:", train_df.shape)
print("Test Shape:", test_df.shape)
```

```python
features = [
    'amt',
    'category',
    'gender',
    'state',
    'lat',
    'long',
    'city_pop',
    'unix_time',
    'merch_lat',
    'merch_long'
]

target = 'is_fraud'
```

```python
categorical_cols = [
    'category',
    'gender',
    'state'
]

encoders = {}

for col in categorical_cols:

    le = LabelEncoder()

    combined = pd.concat([
        train_df[col],
        test_df[col]
    ])

    le.fit(combined)

    train_df[col] = le.transform(train_df[col])

    test_df[col] = le.transform(test_df[col])

    encoders[col] = le
```

```python
X_train = train_df[features]

y_train = train_df[target]

X_test = test_df[features]

y_test = test_df[target]
```

```python
model = RandomForestClassifier(
    n_estimators=100,
    random_state=42
)

model.fit(X_train, y_train)

print("\nModel Training Completed")
```

```python
y_pred = model.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)

print("\n===================================")
print("MODEL ACCURACY")
print("===================================")

print(f"Accuracy: {accuracy*100:.2f}%")

print("\n===================================")
print("CONFUSION MATRIX")
print("===================================")

print(confusion_matrix(y_test, y_pred))

print("\n===================================")
print("CLASSIFICATION REPORT")
print("===================================")

print(classification_report(y_test, y_pred))
```

```python
importance = pd.DataFrame({
    'Feature': features,
    'Importance': model.feature_importances_
})

importance = importance.sort_values(
    by='Importance',
    ascending=False
)

print("\n===================================")
print("FEATURE IMPORTANCE")
print("===================================")

print(importance)
```

```python
print("\n")
print("="*60)
print("      FRAUD DETECTION SYSTEM")
print("="*60)

transaction_no = int(
    input(f"\nEnter Transaction Number (0 - {len(X_test)-1}): ")
)

sample = X_test.iloc[[transaction_no]]

prediction = model.predict(sample)

probability = model.predict_proba(sample)

fraud_probability = probability[0][1] * 100

print("\n")
print("="*60)

if prediction[0] == 1:

    print("FRAUDULENT TRANSACTION")

else:

    print("LEGITIMATE TRANSACTION")

print(f"\nFraud Probability : {fraud_probability:.2f}%")

print("="*60)
```
