```python
import pandas as pd
import numpy as np

from google.colab import drive

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score, classification_report
```

```python
train_data = "/content/drive/MyDrive/archive (2)/Genre Classification Dataset/train_data.txt"

test_data = "/content/drive/MyDrive/archive (2)/Genre Classification Dataset/test_data.txt"

test_data_solution = "/content/drive/MyDrive/archive (2)/Genre Classification Dataset/test_data_solution.txt"
```

```python
print("===================================")
print("SAMPLE DATA FROM TRAIN FILE")
print("===================================")

with open(train_data, "r", encoding="utf-8") as file:

    for i in range(3):
        print(file.readline())
```

```python
train_genres = []
train_plots = []

with open(train_data, "r", encoding="utf-8") as file:

    for line in file:

        # Split dataset properly
        parts = line.strip().split(" ::: ", 3)

        # Ensure proper format
        if len(parts) == 4:

            movie_id = parts[0]
            title = parts[1]
            genre = parts[2]
            plot = parts[3]

            train_genres.append(genre)
            train_plots.append(plot)

# Create dataframe
train_df = pd.DataFrame({
    "genre": train_genres,
    "plot": train_plots
})

print("\n===================================")
print("TRAINING DATAFRAME")
print("===================================")

print(train_df.head())

print("\nTotal Training Samples:", len(train_df))
```

```python
test_plots = []

with open(test_data, "r", encoding="utf-8") as file:

    for line in file:

        parts = line.strip().split(" ::: ", 2)

        if len(parts) >= 2:

            plot = parts[-1]

            test_plots.append(plot)

print("\nTest Plots Loaded:", len(test_plots))
```

```python
test_genres = []

with open(test_data_solution, "r", encoding="utf-8") as file:

    for line in file:

        parts = line.strip().split(" ::: ", 3)

        if len(parts) == 4:

            genre = parts[2]

            test_genres.append(genre)

print("Test Labels Loaded:", len(test_genres))
```

```python
tfidf = TfidfVectorizer(
    stop_words='english',
    max_features=15000,
    ngram_range=(1,2)
)

X_train = tfidf.fit_transform(train_df["plot"])

X_test = tfidf.transform(test_plots)

y_train = train_df["genre"]

y_test = test_genres

print("\nTF-IDF Feature Extraction Completed")
```

```python
model = LinearSVC()

model.fit(X_train, y_train)

print("\nModel Training Completed Successfully")
```

```python
y_pred = model.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)

print("\n===================================")
print("MODEL ACCURACY")
print("===================================")

print(f"Accuracy: {accuracy * 100:.2f}%")

print("\n===================================")
print("CLASSIFICATION REPORT")
print("===================================")

print(classification_report(y_test, y_pred))
```

```python
print("\n===================================")
print("MOVIE GENRE PREDICTION SYSTEM")
print("===================================")

# Take user input
user_plot = input("\nEnter a movie plot:\n\n")

# Convert into TF-IDF features
user_plot_tfidf = tfidf.transform([user_plot])

# Predict genre
predicted_genre = model.predict(user_plot_tfidf)

# Display result
print("\n===================================")
print("PREDICTED GENRE")
print("===================================")

print("Predicted Genre:", predicted_genre[0])
```
